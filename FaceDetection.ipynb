### Coverting to GrayScale

```
#Loading the image to be tested
test_image = cv2.imread('AdityaG.jpg')

#Converting to grayscale
test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)

# Displaying the grayscale image
plt.imshow(test_image_gray, cmap='gray')
```

### Uploading classifier's Xml (We will be using Opencv's pre-trained cascade classifier for face detection)

```
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

### Initializing the classifier for frontal face

```
 cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

#### Using the classifier for detection 

```
faces_rects = cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.2, minNeighbors = 5);

# Peeking into the return values of faces_rects
print(faces_rects)

# no. of faces found
print('Faces found: ', len(faces_rects))
```

```
# it will return a rectangle with coordinates(x,y,w,h) around the face.
for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

#### Showing detected Face

```
#conversion to RGB and show image

plt.imshow(cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB))
```

#### face detection in group

```
#Loading the group image to be tested
test_image2 = cv2.imread('IMG-20161228-WA0000.jpg')

#Converting to grayscale
test_image2_gray = cv2.cvtColor(test_image2, cv2.COLOR_BGR2GRAY)

# Displaying the grayscale image
plt.imshow(test_image2_gray, cmap='gray')
```

```
grp_faces_rects = cascade_face.detectMultiScale(test_image2_gray, scaleFactor = 1.2, minNeighbors =5);

# no. of faces found
print('Faces found: ', len(grp_faces_rects))

# it will return a rectangle with coordinates(x,y,w,h) around the face.
for (x,y,w,h) in grp_faces_rects:
     cv2.rectangle(test_image2, (x, y), (x+w, y+h), (0, 255, 0), 2)

#conversion to RGB and show image

plt.imshow(cv2.cvtColor(test_image2, cv2.COLOR_BGR2RGB))
```

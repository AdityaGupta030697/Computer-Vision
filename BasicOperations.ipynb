```
img = cv2.imread('mandrill.jpg')
print(type(img))
print(img.shape)

# the .jpg image gets transformed into a numpy array with a shape of 288x288 and has 3 color channels.
# OpenCV follows BGR format whereas Matplotlib follows RGB format. Therefore, we change the color order according to Matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### Basic Operations

```
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(blank_img)
```

### Straight Line

```
# Draw a diagonal yellow line with thickness of 10 px
# Syntax cv2.shape(line, rectangle)(image,Point1,Point2,color,thickness)

line = cv2.line(blank_img,(0,0),(511,511),(255,255,0),10)
plt.imshow(line)
```

### Cross

```
line2 = cv2.line(line,(0,511),(511,0),(0,255,0),10)
plt.imshow(line2)
```

### Square

```
#Draw a blue square with a thickness of 5 px
blank_img2 = np.zeros(shape=(512,512,3),dtype=np.int16)
rectangle= cv2.rectangle(blank_img2,(100,100),(400,400),(0,0,255),5)
plt.imshow(rectangle)
```

### Rectangle

```
# Draw a blue rectangle with a thickness of 5 px
# Syntax rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])
# pt1 Vertex of the rectangle. param pt2 Vertex of the rectangle opposite to pt1

blank_img3 = np.zeros(shape=(512,512,3),dtype=np.int16)
rectangle= cv2.rectangle(blank_img3,(100,200),(400,400),(0,0,255),5)
plt.imshow(rectangle)
```

```
blank_img4 = np.zeros(shape=(512,512,3),dtype=np.int16)
circle = cv2.circle(blank_img4,(200,200), 63, (0,0,255), -1) # -1 corresponds to a filled circle
plt.imshow(circle)
```

### Writing on Images

```
font = cv2.FONT_HERSHEY_SIMPLEX
text = cv2.putText(np.zeros(shape=(512,512,3),dtype=np.int16),' Aditya G',(10,300), font, 3,(255,255,255),2,cv2.LINE_AA)
plt.imshow(text)
```

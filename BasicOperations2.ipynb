```
img = cv2.imread('mandrill.png')
print(type(img))
print(img.shape)
cols = img.shape[1]
rows = img.shape[0]

# the .jpg image gets transformed into a numpy array with a shape of 288x288 and has 3 color channels.
# OpenCV follows BGR format whereas Matplotlib follows RGB format. Therefore, we change the color order according to Matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

#### Affine Transformation

```
Matrix = np.float32([[1,0,100],[0,1,100]])
img_shifted = cv2.warpAffine(img,Matrix,(cols,rows))
plt.imshow(img_shifted)
```

#### Rotation

```
center = (cols/2, rows/2)
angle = 45
RMatrix= cv2.getRotationMatrix2D(center,angle,1)
rotated = cv2.warpAffine(img,RMatrix,(cols,rows))
plt.imshow(rotated)
```

#### Conversion to grayscale

```
grayscaled = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(grayscaled)
```

#### Thresholding

```
threshold_value = 125
t_value, binary_threshold = cv2.threshold(grayscaled, threshold_value,255,cv2.THRESH_BINARY)
plt.imshow(binary_threshold)

# Everything that is 125 and under would be converted to 0, or black, and everything above 125 would be converted to 255, or white.
```

## Filtering

### Gaussian Blur

```
matrix = (5,5)

blurred = cv2.GaussianBlur(img, matrix, 0)

plt.imshow(img)
```

```
plt.imshow(blurred)
```

### Median Blur

```
k = 3
median = cv2.medianBlur(img,k)
plt.imshow(img)
```

```
plt.imshow(median)
```
